## Data Preprocessing

Looking at our data, we can see that our data set needs a preprocessing step in order to ensure that the learning algorithms we are going to apply will yield the best results. With this in mind, we are going to apply some changes on the whole data set at first. Afterwards, the data set will be divided into **training** and **test** set, moment at which we will perform an **imputing** operation on the training set since there are missing values.

### 1. Drop unnecessary columns

We have decided that some columns from the data set are not meaningful for our project. Therefore, we decided to completely remove them. These columns represent the wind direction at different moments of time. The columns are:<br>
- WindGustDir
- WindDir9am
- WindDir3pm
- RISK_MM

```python
data_timeframe = data_timeframe.drop(["RISK_MM","RISK_MM_var"],1)
data = data.drop(["WindGustDir"], 1)
data = data.drop(["WindDir9am"], 1)
data = data.drop(["WindDir3pm"], 1)
data = data.drop(["RISK_MM"], 1)
data[0:100]
```

### 2. Date conversion

Considering the fact that, for the moment, we are not making use of timeframes, we have decided to convert the dates into seasons. Giving a value of 1,2,3,4 for each season was not a good aproach because, doing this, we would introduce an ordering relationship. Thus, we have decided to make use of the **one hot encoding** technique.

```python
import numpy as np


def date_to_season(date):
    month = date.split('-')[1]
    return (int(month) % 12 + 3) // 3


def season_one_hot_encoding(season):
    if season == 1:
        return np.array([1, 0, 0, 0])
    if season == 2:
        return np.array([0, 1, 0, 0])
    if season == 3:
        return np.array([0, 0, 1, 0])
    if season == 4:
        return np.array([0, 0, 0, 1])


data['Date'] = data['Date'].apply(date_to_season)
dataFrameResult = pd.DataFrame(data['Date'].apply(season_one_hot_encoding).values.tolist(),
                               columns = ['Winter', 'Spring', 'Summer', 'Autumn'])
data = pd.concat([dataFrameResult, data], sort = False, axis = 1)
data = data.drop(['Date'], 1)

data_timeframe['Date'] = data_timeframe['Date'].apply(date_to_season)
dataTimeFrameResult = pd.DataFrame(data_timeframe['Date'].apply(season_one_hot_encoding).values.tolist(),
                               columns = ['Winter', 'Spring', 'Summer', 'Autumn'])
data_timeframe = pd.concat([dataTimeFrameResult, data_timeframe], sort = False, axis = 1)
data_timeframe = data_timeframe.drop(['Date'], 1)

data[0:100]
```

### 4. Convert yes/no answer

Since we are facing a classification problem, the labels in the dataset which tell if it rains today or tomorrow are strings representing **Yes** or **No** values. Having this in mine, we can easily convert the strings to numerical values, where **Yes** is **1** and **No** is **0**.

```python
def from_answer_to_value(answer):
    if answer == 'Yes':
        return 1
    return 0

data['RainToday'] = data['RainToday'].apply(from_answer_to_value)
data['RainTomorrow'] = data['RainTomorrow'].apply(from_answer_to_value)

data_timeframe['RainToday'] = data_timeframe['RainToday'].apply(from_answer_to_value)
data_timeframe['RainTomorrow'] = data_timeframe['RainTomorrow'].apply(from_answer_to_value)

data[0:100]
```

### 8. Feature correlation

In the data set, there are some features that display information regarding the weather at different moments across the day. These features are not particularly useful as they are, but we can change them in order to display some information regarding the variation of that feature in that timeframe. Thus, we will perform the following reductions:<br>

- Humidity9am and Humidity3pm => Humidity

### 11. Dimensionality reduction

Besides manually correlating features in the dataset, we can make use of the **Principal Component Analysis** algorithm which given a set of unlabeled normalized data, can reduce the size, i.e. the number of features, whilst also maintaining almost the same variance.

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_variance = 0.95

X_train_pca = X_train_raw.copy()
X_test_pca = X_test_raw.copy()

attributes = ['Elevation', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'WindGustSpeed']

scaler = preprocessing.StandardScaler().fit(X_train_pca[attributes])
X_train_pca[attributes] = scaler.transform(X_train_pca[attributes])
X_test_pca[attributes] = scaler.transform(X_test_pca[attributes])

pca = PCA(pca_variance)
X_train_pca = pca.fit_transform(X_train_pca)
X_test_pca = pca.transform(X_test_pca)

X_train_pca = np.asarray(X_train_pca)
X_test_pca = np.asarray(X_test_pca)

X_train_pca = pd.DataFrame(X_train_pca, index = [x for x in range(X_train.shape[0])], columns = [x for x in range(len(X_train_pca[0]))])
X_test_pca = pd.DataFrame(X_test_pca, index = [x for x in range(X_test.shape[0])], columns = [x for x in range(len(X_test_pca[0]))])

print("Number of features returned by applying PCA while maintaining a variance of " + str(pca_variance) + " is: " + str(X_train_pca.shape[1]))
X_train_pca[0:100]
```

## Supervised Learning Models

In the following section, we will apply different classification learning algorithms. The algorithms will be applied on **3** different sets:<br>
- Raw data
- Preprocessed data (the preprocessing done by us)
- PCA data
- Timeframe data (preprocessed data which also contains the variation of some features between today and yesterday)

For each set, we have made use of the __StratifiedKFold__ library from python in order to test our learning models against a cross validation set. The cross validation set represents __10%__ of the data set, with the remaining __90%__ being used for fitting our models (the process is repeated 10 times for each set; the cross validation and the training set are splitted randomly). The result after applying each algorithm represents the classification metrics that we have previoulsy mentioned.

### 1. Logistic Regression

The first model we are going to use is **Logistic Regression**, which is a widely used statistical model that, in its basic form, uses a logistic function to model a binary dependent variable. In regression analysis, logistic regression is estimating the parameters of a logistic model. This model is already defined in the **sklearn** library. The configuration parameters that we have used are:<br>
- _tol_: the tolerance value (the value at which it is considered that convergence has occured) is __1e-4__
- _C_: the inverse of regularization strength is __1.0__ (smaller values specify stronger regularization)
- _random state_: the seed of the pseudo-random number generator to use when shuffling the data; constant in order to preserve the same behavior amongst different executions
- _solver_: the algorithm used in the optimization problem. We use __Stochastic Average Gradient__ because it ensures fast convergence on large datasets with features being approximately on the same scale
- _max iter_: the maximum number of iterations taken for the solvers to converge
- _n jobs_: the number of cpu cores to be used. **-1** means all

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold

metrics_list = ['precision', 'recall', 'f1-score']
sets = ['train', 'cross']

def logistic_regression(X_train, y_train, degree = 1, cv_size = 10, type = 0):

    global log_regr_pca_model 
    global log_regr_raw_model
    global log_regr_preprocessed_model
    global log_regr_timeframe_model
    
    report = {}
    for set in sets:
        report[set] = {}
        for metric in metrics_list:
            report[set][metric] = 0
    
    random_state = 12883823
    skf = StratifiedKFold(n_splits = cv_size)

    for train, cross in skf.split(X_train, y_train):
        x_train_set = X_train.iloc[train]
        y_train_set = y_train.iloc[train]
        x_cross_set = X_train.iloc[cross]
        y_cross_set = y_train.iloc[cross]
        
        polynomial_degree = PolynomialFeatures(degree = degree)
        polynomial_training = polynomial_degree.fit_transform(x_train_set)
        
        logistic_regr = LogisticRegression(C = 1,solver = 'sag', max_iter = 300, n_jobs = -1, random_state = random_state).fit(polynomial_training, y_train_set)
    
        train_predicted = logistic_regr.predict(polynomial_training)
        cv_predicted = logistic_regr.predict(polynomial_degree.fit_transform(x_cross_set))
    
        train_metric = metrics.classification_report(y_train_set, train_predicted, output_dict = True)
        cv_metric = metrics.classification_report(y_cross_set, cv_predicted, output_dict = True)
        
        for metric in metrics_list:
            report['train'][metric] += train_metric['1'][metric]
            
        for metric in metrics_list:
            report['cross'][metric] += cv_metric['1'][metric]
            
        if type == 0:
            log_regr_pca_model = logistic_regr
        if type == 1:
            log_regr_preprocessed_model = logistic_regr
        if type == 2:
            log_regr_raw_model = logistic_regr
        if type == 3:
            log_regr_timeframe_model = logistic_regr
            
        
        
    for set in sets:
        for metric in metrics_list:
            report[set][metric] /= cv_size
    
    return report
    

report_pca = logistic_regression(X_train_pca, y_train, type = 0)
report_preprocessed = logistic_regression(X_train, y_train, type = 1)
report_raw = logistic_regression(X_train_raw, y_train, type = 2)
report_timeframe = logistic_regression(X_timeframe_train, y_timeframe_train, type = 3)

report_log_regr= pd.DataFrame()
report_log_regr['Index'] = ['Precision', 'Recall', 'F1-score']
report_log_regr.set_index('Index', inplace = True)
report_log_regr['Raw'] = report_raw['cross'].values()
report_log_regr['PCA'] = report_pca['cross'].values()
report_log_regr['Preprocessed'] = report_preprocessed['cross'].values()
report_log_regr['Timeframe'] = report_timeframe['cross'].values()

report_log_regr
```

### 2. Neural  Network

The next model we are going to apply is **Neural Networks**. An Artificial Neural Network (ANN) is an information processing paradigm that is inspired by the way biological nervous systems, such as the brain, process information. The key element of this paradigm is the novel structure of the information processing system. It is composed of a large number of highly interconnected processing elements (neurones) working in unison to solve specific problems. ANNs, like people, learn by example. Learning in biological systems involves adjustments to the synaptic connections that exist between the neurones.

This model is already defined in the **sklearn** library. The configuration parameters that we have used are:<br>
- _hidden layer sizes_: a tuple where the **ith** element represents the number of neurons in the **ith** hidden layer. We have chosen to have only one hidden layer having a number of neurons equal to half the number of features
- _activation_: activation function for the hidden layer. We have chosen to use the **logistic sigmoid function**
- _solver_: the solver for weight optimization. We have chosen **adam** which is a stochastic gradient-based optimizer because it works well on large datasets.
- _alpha_: the regularization term; its value is **e1-4**
- _batch_size_: size of minibatches for stochastic optimizers; its value is **200**
- _max iter_: the maximum number of iterations taken for the solvers to converge
- _shuffle_: wheter to shuffle samples in each iteration; set to **True**
- _random state_: the seed of the pseudo-random number generator to use when shuffling the data; constant in order to preserve the same behavior amongst different executions
- _tol_: the tolerance value (the value at which it is considered that convergence has occured) is __1e-4__

### 3. Support Vector Machine

The next model we are going to use is **Support Vector Machine**. A Support Vector Machine (SVM) is a discriminative classifier formally defined by a separating hyperplane. In other words, given labeled training data (supervised learning), the algorithm outputs an optimal hyperplane which categorizes new examples. In two dimentional space this hyperplane is a line dividing a plane in two parts where in each class lay in either side.

This model is already defined in the **sklearn** library. The configuration parameters that we have used are:<br>
- _C_: the inverse of regularization strength is __1.0__ (smaller values specify stronger regularization)
- _kernel_: specified the kernel type to be used in the algorithm; we have made use of **Radial basis function kernel**
- _gamma_: defines how far the influence of a single training example reaches, with low values meaning ‘far’ and high values meaning ‘close’; its value is **scale** 
- _tol_: the tolerance value (the value at which it is considered that convergence has occured) is __1e-3__
- _max iter_: the maximum number of iterations taken for the solvers to converge
- _random state_: the seed of the pseudo-random number generator to use when shuffling the data; constant in order to preserve the same behavior amongst different executions


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold

metrics_list = ['precision', 'recall', 'f1-score']
sets = ['train', 'cross']

def svm_classifier(X_train, y_train, degree = 1, cv_size = 10, type = 0):

    global svm_preprocessed_model
    global svm_raw_model
    global svm_pca_model
    global svm_timeframe_model
    
    report = {}
    for set in sets:
        report[set] = {}
        for metric in metrics_list:
            report[set][metric] = 0
    
    random_state = 12883823
    skf = StratifiedKFold(n_splits = cv_size)

    for train, cross in skf.split(X_train, y_train):
        x_train_set = X_train.iloc[train]
        y_train_set = y_train.iloc[train]
        x_cross_set = X_train.iloc[cross]
        y_cross_set = y_train.iloc[cross]
        
        polynomial_degree = PolynomialFeatures(degree = degree)
        polynomial_training = polynomial_degree.fit_transform(x_train_set)
        
        svm = SVC(C = 1.0, kernel = 'rbf', gamma = 'scale', max_iter = 1200, random_state = random_state).fit(polynomial_training, y_train_set)
    
        train_predicted = svm.predict(polynomial_training)
        cv_predicted = svm.predict(polynomial_degree.fit_transform(x_cross_set))
    
        train_metric = metrics.classification_report(y_train_set, train_predicted, output_dict = True)
        cv_metric = metrics.classification_report(y_cross_set, cv_predicted, output_dict = True)
                
        for metric in metrics_list:
            report['train'][metric] += train_metric['1'][metric]
            
        for metric in metrics_list:
            report['cross'][metric] += cv_metric['1'][metric]
        
        if type == 0:
            svm_pca_model = svm
        if type == 1:
            svm_preprocessed_model = svm
        if type == 2:
            svm_raw_model = svm
        if type == 3:
            svm_timeframe_model = svm
        
        
    for set in sets:
        for metric in metrics_list:
            report[set][metric] /= cv_size
    
    return report
    

report_pca = svm_classifier(X_train_pca, y_train, type = 0)
report_preprocessed = svm_classifier(X_train, y_train, type = 1)
report_raw = svm_classifier(X_train_raw, y_train, type = 2)
report_timeframe = svm_classifier(X_timeframe_train, y_timeframe_train, type = 3)

report_svm = pd.DataFrame()
report_svm['Index'] = ['Precision', 'Recall', 'F1-score']
report_svm.set_index('Index', inplace = True)
report_svm['Raw'] = report_raw['cross'].values()
report_svm['PCA'] = report_pca['cross'].values()
report_svm['Preprocessed'] = report_preprocessed['cross'].values()
report_svm['Timeframe'] = report_timeframe['cross'].values()

report_svm
```

### 4. Decision Trees

The last model we have used is **Decision Trees**. Decision trees are the most important elements of a Random Forest. They are capable of fitting complex data sets while allowing the user to see how a decision was taken. They are already implemented in the **sklearn** library.The configuration parameters are the following:<br>
- _criterion_: the function to measure the quality of a split; we have chosen the **gini** impurity
- _splitter_: the strategy used to choose the split at each node; we have chosen the **best** split
- _max depth_: the maximum depth of the tree; we have chosen to not impose it
- _random state_: the seed of the pseudo-random number generator to use when shuffling the data; constant in order to preserve the same behavior amongst different executions
- _max features_: the number of features to consider when looking for the best split; we have chosen to consider all features

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold

metrics_list = ['precision', 'recall', 'f1-score']
sets = ['train', 'cross']


def decision_tree_classifier(X_train, y_train, degree = 1, cv_size = 10, type = 0):
    
    global decision_tree_preprocessed_model
    global decision_tree_raw_model
    global decision_tree_pca_model
    global decision_tree_timeframe_model

    report = {}
    for set in sets:
        report[set] = {}
        for metric in metrics_list:
            report[set][metric] = 0
    
    random_state = 12883823
    skf = StratifiedKFold(n_splits = cv_size)

    for train, cross in skf.split(X_train, y_train):
        x_train_set = X_train.iloc[train]
        y_train_set = y_train.iloc[train]
        x_cross_set = X_train.iloc[cross]
        y_cross_set = y_train.iloc[cross]
        
        polynomial_degree = PolynomialFeatures(degree = degree)
        polynomial_training = polynomial_degree.fit_transform(x_train_set)
        
        decision_tree = DecisionTreeClassifier(max_depth = 3, random_state = random_state).fit(polynomial_training, y_train_set)
    
        train_predicted = decision_tree.predict(polynomial_training)
        cv_predicted = decision_tree.predict(polynomial_degree.fit_transform(x_cross_set))
    
        train_metric = metrics.classification_report(y_train_set, train_predicted, output_dict = True)
        cv_metric = metrics.classification_report(y_cross_set, cv_predicted, output_dict = True)
        
        for metric in metrics_list:
            report['train'][metric] += train_metric['1'][metric]
            
        for metric in metrics_list:
            report['cross'][metric] += cv_metric['1'][metric]
        
        if type == 0:
            decision_tree_pca_model = decision_tree
        if type == 1:
            decision_tree_preprocessed_model = decision_tree
        if type == 2:
            decision_tree_raw_model = decision_tree
        if type == 3:
            decision_tree_timeframe_model = decision_tree
        
        
    for set in sets:
        for metric in metrics_list:
            report[set][metric] /= cv_size
    
    return report
    

report_pca = decision_tree_classifier(X_train_pca, y_train, type = 0)
report_preprocessed = decision_tree_classifier(X_train, y_train, type = 1)
report_raw = decision_tree_classifier(X_train_raw, y_train, type = 2)
report_timeframe = decision_tree_classifier(X_timeframe_train, y_timeframe_train, type = 3)

report_decision_tree = pd.DataFrame()
report_decision_tree['Index'] = ['Precision', 'Recall', 'F1-score']
report_decision_tree.set_index('Index', inplace = True)
report_decision_tree['Raw'] = report_raw['cross'].values()
report_decision_tree['PCA'] = report_pca['cross'].values()
report_decision_tree['Preprocessed'] = report_preprocessed['cross'].values()
report_decision_tree['Timeframe'] = report_timeframe['cross'].values()

report_decision_tree
```
